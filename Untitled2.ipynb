```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 設定台積電的股票代碼 (TSM 是在紐約證券交易所上市的代碼)
ticker_symbol = 'TSM'

# 設定獲取數據的日期範圍
start_date = '2022-01-01'
end_date = pd.to_datetime('today').strftime('%Y-%m-%d')

# 下載歷史股價數據
data = yf.download(ticker_symbol, start=start_date, end=end_date)

# 顯示數據的前幾行，確認數據已正確載入
print(f'{ticker_symbol} 股價數據 (前5行):')
display(data.head())
```

```python
# 繪製台積電的收盤價走勢圖
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='TSM Close Price', color='blue')
plt.title(f'{ticker_symbol} 股價走勢圖 ({start_date} 至 {end_date})')
plt.xlabel('日期')
plt.ylabel('收盤價 (USD)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
